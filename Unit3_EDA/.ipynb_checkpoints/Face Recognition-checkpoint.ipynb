```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from random import randint
import matplotlib.cm as cm
from skimage import io, color
import os
from PIL import Image
from IPython.display import clear_output
```

```python
# %% Read image data
numImg = 20
numSbj = 12
fPath = 'faces94' #Path to face94 database
j = numSbj
c = 0
A = pd.DataFrame([])
y = pd.DataFrame([])
for i in os.listdir(fPath):
    if(j <= 0):
        break
    j -= 1
    for f in os.listdir(fPath + '/' + i):
        imgPath = fPath + '/' + i + '/' + f
        for filename in os.listdir(imgPath):
            if filename.endswith('.jpg'):
                try:
                    B = pd.DataFrame(color.rgb2gray(io.imread(imgPath+'/'+filename)).reshape([1,180*200])).reset_index(drop=True)
                    A = pd.concat([A,B])
                    z = pd.DataFrame([j])
                    y = pd.concat([y,z])
                    c = c + 1
                    clear_output()
                    print(i,j,c,filename)
                except(ValueError,IOError,SyntaxError) as e:
                    print(e)
                    
A = A.reset_index(drop=True)
y = y.reset_index(drop=True)
```

```python
print(A.shape,y.shape)
```

```python
plt.figure()
pca = PCA(n_components=4)
pca.fit(A)
for i in range(4):
    ax = plt.subplot(2, 2, i + 1)
    ax.imshow(pca.components_[i].reshape([200, 180]))
plt.show()
```

```python
#Colors for distinct individuals
cols = ['#{:06x}'.format(randint(0, 0xffffff)) for i in range(numSbj)]
pltCol = [cols[int(k)] for k in y]
# %% Plot 2d PCA data
drA = pca.transform(A)
plt.figure()
plt.scatter(drA[:, 0], drA[:, 1], color=pltCol)
plt.show()
```
