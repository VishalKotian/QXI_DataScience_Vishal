```python
img = cv2.imread('DATA/00-puppy.jpg')
```

```python
img
```

```python
plt.imshow(img)
# since cv2 reads an image in BGR format not in RGB as matplotlib
```

The image has been correctly loaded by openCV as a numpy array, but the color of each pixel has been sorted as BGR. Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels. This operation can be done either by using openCV conversion functions cv2.cvtColor() or by working directly with the numpy array.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# to convert the color from BGR to RGB
```

```python
plt.imshow(img_rgb)
```

```python
# to show the image in Grayscale
img_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

## Resize Images

```python
img_rgb.shape
# height, width, color channel
```

```python
img = cv2.resize(img_rgb,(1300,275)) # where 1300 is the width and 275 is the height
```

```python
plt.imshow(img)
```

## To reduce size by ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb,(0,0), img_rgb, w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

## Flipping the images

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```

## Saving Image Files

```python
type(new_img)
```

```python
cv2.imwrite('my_new_picture.jpg',new_img)
#Keep in mind, the above stored the BGR version of the image.
```
